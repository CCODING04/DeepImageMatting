```
import os
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import cv2 as cv
%matplotlib inline
```

```
imgs = list()
aplhas = list()
```

```
figure, cells = plt.subplots(1,5, figsize = (16,8))
for i in range(5):
  img = cv.imread('color_{0}.jpg'.format(i+1))[:,:,[2,1,0]]
  imgs.append(img)
  cells[i].imshow(img)
  cells[i].axis('off')
```

```
figure, cells = plt.subplots(1,5, figsize = (16,8))
for i in range(5):
  img = cv.imread('alpha_{0}.jpg'.format(i+1))[:,:,[2,1,0]]
  imgs.append(img)
  cells[i].imshow(img)
  cells[i].axis('off')
```
